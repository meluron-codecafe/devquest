Think of virtual environment as a **special folder/directory** that stores all the external python tools you need for a project. This helps isolate your project related tools from other projects or from system-wide available tools. \
You can name the folder anything you want and have it located anywhere on your machine but a general convention is to have virtual environment folder for a project in the project root directory by the name **.venv**

```python
from modusa.generators import AudioWaveformGenerator
```

```python
from IPython.display import display, clear_output
import time

for f in range(10, 20, 1):
    clear_output(wait=True)
    sine = AudioWaveformGenerator.generate_sinusoid(f=f, sr=1000)
    saw =  AudioWaveformGenerator.generate_sawtooth(f=f, phi=1, sr=1000)
    display((sine + saw).plot(title=f"{f} hz", ylim=(-3, 3)))
    time.sleep(0.1)
```

```python
sine = AudioWaveformGenerator.generate_sinusoid(f=800, sr=10000, duration=5)
saw = AudioWaveformGenerator.generate_sawtooth(A=0.5,f=1005, phi=0.4, sr=10000, duration=5)
square = AudioWaveformGenerator.generate_square(A=0.5,f=1000, phi=0.4, sr=10000, duration=10).crop(5, 10)
y = sine + saw + square
y.plot(xlim=(0, 0.02))
```

```python
y.crop(0, 0.02).plot(stem=True)
```
