## Signal
A **signal** is anything that contains **information** and varies over time or location. This **variation** is what encapsulates information. If the signal is not varying, e.g. constant signal, there is not much information to learn from it.

A **signal** is **continuous** in nature, i.e. there is a value at every instant of time/space. However, a proof might be needed here.

Since a signal encapsulates behaviour of a system so analsing the signal becomes really important to understand the system. One can hypothetically analyse signal manually, by analysing values one-by-one but we do have computers to handle that tedious task of performing operations that too with better accuracy, speed and efficiency.

But there is a catch, in order to use *computers*, we have to store the signal into **memory** that computers can access. Since, signals that we deal with are continuous, there is no way to store that infinite amount of information in computer. There are various **quantization** that we have to perform to allow computers to store signal. Signal thus stored in computer is referred to as **digital signal**.

## ADC
The process of converting an analog/continuous signal into digital/discrete signal is called **ADC**. It stands for **Analog to Digital Conversion**. This process involves **quantization** of the signal so that they can be stored in computer.

### Time Quantization / Sampling

Since we have limited memory to store signal in computer, we perform *time quantization* which means that we only collect values of the analog signal at certain timepoints often equally spaced by say $T$ sec which is known as **sampling period** of the signal. The inverse of $T$ is called **sampling frequency** $f_s = \frac{1}{T}$ hz

```python
#hide
import numpy as np
import modusa as ms

pi = np.pi
f = 2
sr = 100
dur = 1

t = np.arange(int(sr*dur)) / sr
analog_y = np.sin(2*pi*f*t)

display(ms.plot1d(
    (analog_y, t),
    ylabel="Amplitude",
    legend="Analog Signal"
))

display(ms.plot1d(
    (np.ones(t.size), t),
    title="X",
    ylabel="Amplitude",
    xlabel="Time (sec)",
    legend="Pulse Train",
    show_stem=True,
))

display(ms.plot1d(
    (analog_y, t),
    title="=",
    ylabel="Amplitude",
    xlabel="Time (sec)",
    legend="Digital",
    show_stem=True
))
```

### Amplitude Quantization

After performing *time quantization*, we still have to quantize the values as computers cannot store values with infinite precision. We first decide something called **bit depth** (number of bits per sample), which is +ve integer value often multiples of 2. \
$bit depth = n \implies L = 2^{16}$ \
where $L$ is the number of levels available for each sample of the signal.

All the values of the analog signal are rounded to the nearest bin computed as per the bit-depth.

The range of numbers for a given bit depth depends on the datatype used to store the values.

Due to this amplitude quantization, we get **quantisation error** leading to **quantization noise**
This is often reported as sound-to-quantization-noise-ratio aka $SQNR$.
$SQNR \propto n $
