When we run python code from terminal, we can activate any virtual environment and use it directly without any issues.

However while using Jupyter, it is important to understand that it has its own way to allow virtual environments to be used. This is through what we call as kernels. Think of kernel as a configuration that allows Jupyter to access different existing virtual environments.

For example, we see a list of kernels available to us in the Jupyter notebooks below.

<img src="../data/images/jupyter-lab-launcher.png" width=600 height=300>

````python
#hide
utils.h1("What is a kernel?")
````

A kernel for a Jupyter notebook is a way to point to the virtual environment so that it can run the codeblock using a virtual env.

````python
#hide
utils.h1("Install a kernel using ipykernel module")
````

```python -m ipykernel install --user --name=<kernel_name> --display-name=<name to show in jupyter>```

````python
#hide
utils.h1("How to list all the installed kernels?")
````

jupyter kernelspec list

````python
utils.h1("How to list remove an installed kernel?")
````

jupyter kernelspec remove \<kernel_name>

Just open the kernels folder and delete the kernel.
