## pYIN

**Probablisitic YIN** aka **pYIN** is a fundamental frequency tracking algorithm.

```python
#hide
import modusa as ms
import numpy as np
```

```python
y, sr, title = ms.load("https://www.youtube.com/watch?v=roVPyzV-TPE", sr=44100)
```

```python
ms.plot1d(
    (y, np.arange(y.size) / sr),
    title=title,
    xlabel="Time (s)",
    ylabel="Amplitude",
    legend="Audio Signal"
)
```
