I welcome you to this quest where I am trying to create a **TTS** (Text to speech) model from scratch with just my basic knowledge of audio signal processing. At the time of writing this, I literally have no idea on how to create one but the problem is really interesting.

The easiest way to create a text to speech system that I can think of is by recording audio for different words in the vocabulary and then playing it back as per the text input. Sure, let's build it then.

To keep things simple and with small vocabulary, I will create a text to speech system for reciting a given phone number.

```python
import numpy as np
import librosa
import IPython.display as ipd
```

```python
# Load audio with all the spoken numbers 
audio, sr = librosa.load("../data/audio/count.wav", sr=16000)
ipd.Audio(audio, rate=sr)
```

```python
# Load labels mentioning the audio segment corresponding to each number
with open("../data/audio/count_labels.txt", "r") as f:
    content = [content.split("\t") for content in f.read().split("\n")][:-1]
content[0]
```

```python
def get_audio_segment(number: int):
    start, end = float(content[number][0]), float(content[number][1])
    audio_segment = audio[int(start*sr):int((end + 0.3)*sr)]
    
    return audio_segment
```

```python
phone_number_audio = []

def recite_phone_number(phone_number: str) -> np.ndarray:
    for digit in phone_number:
        audio_segment = get_audio_segment(number=int(digit))
        phone_number_audio.extend(audio_segment)
    return phone_number_audio
```

```python
final_audio = recite_phone_number("911923849")
```

```python
ipd.Audio(final_audio, rate=sr)
```
