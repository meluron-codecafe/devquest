```python
from modusa.signals.signal1d import Signal1D
from modusa.signals.signal2d import Signal2D
from modusa.signals.time_domain_signal import TimeDomainSignal
import numpy as np
import matplotlib.pyplot as plt
```

```python
signal1 = Signal2D.from_array(np.random.random((10, 20)))
signal2 = Signal1D(y=np.random.random(10))
signal1.set_plot_labels(title="Matrix")
signal1.set_units(M_unit="dB")
```

```python
fig, axs = plt.subplots(3, 2, figsize=(14, 6), width_ratios=[1, 0.02], height_ratios=[1, 0.2, 0.2])

# Share x-axis manually between first column (axs[0,0] and axs[1,0])
axs[1, 0].sharex(axs[0, 0])  # Now only first column shares x-axis
axs[2, 0].sharex(axs[0, 0])  # Now only first column shares x-axis


# Plot
signal1.plot(ax=axs[0, 0], show_colorbar=True, cax=axs[0, 1], origin="lower")
signal2.plot(ax=axs[1, 0], color="r")

# Clean up unused colorbar slot in bottom row
axs[1, 1].remove()
axs[2, 1].remove()

plt.tight_layout()
plt.show()
```
