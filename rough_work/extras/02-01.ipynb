**Tempo** refers to the **speed** of a song and measured by how quick **events** occur. These **events** can be literally anything but generally we stick to the percussive hit (kick, even guitar/keys stroke). For e.g., How fast the kick of a drum is being played. Here the kick onset becomes the event. \
Whenever we hear a song, we tend to tap at a certain tempo which is often influenced by these percussive hits.

The **goal** here is to measure the rate of these percussive hits which will eventually tell us the tempo at which we should ideally tap while listening to the song.
