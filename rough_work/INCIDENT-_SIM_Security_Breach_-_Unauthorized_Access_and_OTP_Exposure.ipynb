## 11:02 (17th Sep, 2025)

At 11:02 AM, I received a message that significantly escalated the situation. It stated that 55.83 NeuCoins (in-app reward points used for discounts) had been spent from my account.

## 11:30 (17th Sep, 2025)

I was initially unaware of the full extent of the issue. To investigate, I downloaded the Tata Neu app and logged into my account, where I discovered that an order had been placed from Visakhapatnam, a location where I have no relatives or friends who could have used my credentials.

At this stage, I was uncertain whether the issue stemmed from the app itself or from something more serious. I contacted Tata Neu customer support to report the matter. Their immediate response was to log me out from all devices and prevent any return orders from being placed. I requested to be connected with the person who had placed the order, but they declined due to privacy policies. I then asked if I could speak with the delivery person to understand how the recipient was contacted, since my own number was listed in the app. If the delivery call had been directed to the number shown in the app, I should have received it unless someone had unauthorized access to my SIM and was able to intercept or redirect calls. The support team informed me that the delivery person had successfully contacted the recipient and completed the delivery. This revelation left me shocked and without answers, as the only information I had was the recipient’s name and address from the invoice.

It is worth noting that, by this time, the app displayed my contact number, which created significant confusion. If the app was showing my number, how was the delivery person’s call being received by someone else?

## 12:15 (17th Sep, 2025)

At this point, I began receiving multiple OTPs for Tata Neu. My assumption is that when customer support logged me out from all devices, the recipient was also logged out. By coincidence, I logged back into the app at the same time as the other party. It appears he had already changed the registered contact number to his own, likely in an attempt to place a return order. In that moment, I noticed that the contact number displayed in the app was no longer mine, and the associated email address was also fraudulent. I captured a screenshot of this evidence, with the recipient’s number blurred out for privacy reasons.

The person managed to obtain a product worth ₹760 for free by using my account’s wallet balance, a new-user promo code, and some Tata Neu reward points. In fact, had the NeuCoins not been used, I would likely not have received any notification that prompted me to investigate further.
